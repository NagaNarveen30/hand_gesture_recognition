```python
import os
import yaml
from ultralytics import YOLO

# 1. Resolve paths inside the cloned repo
BASE_DIR  = os.getcwd() 
DATA_DIR  = os.path.join(BASE_DIR, "dataset")
DATA_YAML = os.path.join(DATA_DIR, "data.yaml")

# 2. Update the YAML 'path' field so YOLO knows where to look
with open(DATA_YAML, "r") as f:
    data_cfg = yaml.safe_load(f)

data_cfg["path"] = DATA_DIR

with open(DATA_YAML, "w") as f:
    yaml.dump(data_cfg, f)

# 3. Load pretrained YOLOv8 nano model
model = YOLO("yolov8n.pt")

# 4. Train
# Using 'device=0' ensures it uses the GPU
results = model.train(
    data=DATA_YAML,
    epochs=50,
    imgsz=640,
    batch=16,
    name="gesture_model",
    patience=10,
    device=0 
)

# 5. Evaluate
metrics = model.val()
print(f"mAP50: {metrics.box.map50:.4f}")
```
